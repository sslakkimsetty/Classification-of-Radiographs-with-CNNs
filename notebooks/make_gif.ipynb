```
walk_gen = list(os.walk('cam_images'))
parts = walk_gen[0][1]
parts
```

```
ids = list(range(1,26,4))
```

```
for i, part in enumerate(parts):
    print("########\nFor part,", part.upper())
    idx = ids[i]
    
    dpath = walk_gen[idx][0]
    patients = walk_gen[idx][1]
    
    for patient in patients:
        dir_ = dpath + '/' + patient + '/'
        images = os.listdir(dir_)
        images.sort()
        print(patient)
        
        for img_name in images:
            img_dir = dir_ + img_name
            
            image = Image.open(img_dir)
            draw = ImageDraw.Draw(image)
            text1 = "Patient: " + patient
            text2 = "Epoch: " + img_name.split('.')[0]
            text = text1 + '\n' + text2
            
            draw.text((115, 0), text, fill='red')
            image.save(img_dir)
```
